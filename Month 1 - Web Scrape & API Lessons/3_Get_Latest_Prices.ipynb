```python
import pandas as pd

# Load the DataFrames
df = pd.read_csv('exports/stocks.csv')
df2 = pd.read_csv('exports/historicalprices.csv')

# Assuming your DataFrame is named 'df'
# First, ensure that the 'Date' column is in datetime format
df2['Date'] = pd.to_datetime(df2['Date'])

# Then, group the DataFrame by 'stock_code' and find the maximum date in each group
latest_dates = df2.groupby('stock_code')['Date'].max().reset_index()

# Add one day to the maximum date
latest_dates['Date'] = latest_dates['Date'] + pd.DateOffset(days=1)

# Print or work with the resulting DataFrame containing the latest dates for each 'stock_code'
print(latest_dates)
```

```python
# Convert the 'stock_code' column to strings and add ".KL" to each value
latest_dates['stock_code'] = latest_dates['stock_code'].astype(str).apply(lambda x: x + ".KL")
```

```python
latest_dates
```

```python
import pandas as pd
import yfinance as yf
import datetime

# Assuming you have already loaded your 'latest_dates' DataFrame

# Initialize empty lists to store successful and unsuccessful symbols
combined_data = []
error_symbols = []

# Define the end date as the present day (you may need to import datetime)
end_date = pd.to_datetime('today').strftime('%Y-%m-%d')

# Loop through each stock symbol and retrieve its historical data
for _, row in latest_dates.iterrows():
    symbol = row['stock_code']
    latest_date = row['Date']
    
    try:
        # Use yfinance to retrieve historical data for the symbol
        stock_data = yf.download(symbol, start=latest_date, end=end_date)
        
        # Create a new DataFrame with 'stock_data' and a 'stock_code' column
        stock_data['stock_code'] = symbol
        
        # Append the DataFrame to the list of successful symbols
        combined_data.append(stock_data)
        
        # Print the first few rows of the historical data for demonstration
        print(f"Stock Symbol: {symbol}")
        print(stock_data.head())
        
    except Exception as e:
        print(f"Error retrieving data for {symbol}: {e}")
        # Append the unsuccessful symbol to the list
        error_symbols.append(symbol)

# After the loop, concatenate all DataFrames into a single DataFrame
final_df = pd.concat(combined_data)

# The 'final_df' DataFrame contains the combined historical data for all successful stock symbols
```

```python
final_df
```

```python
# Specify the file path where you want to save the CSV file
csv_file_path = 'exports/newhistoricalpricescheck.csv'

# Use the to_csv() method to export the DataFrame to a CSV file
final_df.to_csv(csv_file_path, index=True)  # Set index=False to exclude the DataFrame index from the CSV

print(f"DataFrame saved to {csv_file_path}")
```

```python
# Removing the last three characters (.KL)
df2['stock_code']=df2['stock_code'].str.slice(0,-3)
df2
```

```python
# Removing unneeded columns
df2.drop(['Adj Close', 'Volume'], axis=1, inplace=True)
df2
```

```python
# Round to 3 decimal places
decimal_places = 3  # Number of decimal places to round to

df2['Open'] = df2['Open'].round(decimal_places)
df2['High'] = df2['High'].round(decimal_places)
df2['Low'] = df2['Low'].round(decimal_places)
df2['Close'] = df2['Close'].round(decimal_places)

df2
```

```python
# Specify the file path where you want to save the CSV file
csv_file_path = 'exports/newhistoricalprices.csv'

# Use the to_csv() method to export the DataFrame to a CSV file
df2.to_csv(csv_file_path, index=False)  # Set index=False to exclude the DataFrame index from the CSV

print(f"DataFrame saved to {csv_file_path}")
```
